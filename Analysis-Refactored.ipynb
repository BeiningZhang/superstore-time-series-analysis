# Import data

```python
def load_data(file_path):
    """
    Load data from Excel file and return DataFrame.
    
    Parameters:
    file_path (str): Path to the Excel file.
    
    Returns:
    pandas.DataFrame: DataFrame containing the loaded data.
    """
    try:
        return pd.read_excel(file_path)
    except FileNotFoundError:
        print("File not found. Please check the file path.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

df = load_data("superstore.xls")
df.head(2)
```

# Data Preprocessing
- Filter data for a specific category
- Remove unnecessary columns (only keep `Order Date` and `Sales`)
- Aggregate sales by date
- Reindex with Order Date
- Resampling to monthly frequency

```python
def preprocess_data(df, category):
    """
    Preprocess data by filtering for a specific category, aggregating sales by date, and resampling to monthly frequency.
    
    Parameters:
    df (pandas.DataFrame): Input DataFrame containing sales data.
    category (str): Category to filter the data.
    
    Returns:
    pandas.Series: Preprocessed time series data.
    """
    if category not in df['Category'].unique():
        print(f"Category '{category}' not found in the data.")
        return None
    
    filtered_df = df[df['Category'] == category]
    aggregated_sales = filtered_df.groupby('Order Date')['Sales'].sum().reset_index()
    aggregated_sales = aggregated_sales.set_index('Order Date')['Sales'].resample('MS').mean()
    
    return aggregated_sales


furniture = preprocess_data(df, 'Furniture')
office_supplies = preprocess_data(df, 'Office Supplies')
```

Our current data is aggregated to single days, which can be tricky to work with because the dates are not successive. Instead, we resample the data to monthly and use average daily sales for each month. We use the start of each month as the timestamp.

# Train-test split

```python
def train_test_split(data, train_end_date):
    """
    Split data into train and test sets based on the specified end date for the training data.
    
    Parameters:
    data (pandas.Series): Time series data to split.
    train_end_date (str): End date for the training data (default: '2016-12-31').
    
    Returns:
    pandas.Series, pandas.Series: Training and testing sets.
    """
    try:
        train = data[:train_end_date]
        test = data[train_end_date:]
        return train, test
    except KeyError:
        print("Invalid date format or out-of-range date. Please provide a valid date format (YYYY-MM-DD).")
        return None, None


furniture_train, furniture_test = train_test_split(furniture, train_end_date='2016-12-31')
office_supplies_train, office_supplies_test = train_test_split(office_supplies, train_end_date='2016-12-31')
```

# Time Series Decomposition

We can also visualize our data using a time series decomposition that allows us to decompose our time series into four distinct components: **level, trend, seasonality, and noise**. A time series can be broke into Systematic Components (have consistency or recurrence and can be described and modeled) and Non-Systematic Components (cannot be directly modeled). A given time series is thought to consist of three systematic components including level, trend, seasonality, and one non-systematic component called noise.

These components are defined as follows:
- Level: The average value in the series.
- Trend: The increasing or decreasing value in the series.
- Seasonality: The repeating short-term cycle in the series.
- Noise: The random variation in the series.

`statsmodels.tsa` is a module within the StatsModels library in Python, which stands for "Time Series Analysis." StatsModels provides utilities for plotting time series data, decomposing time series into trend, seasonal, and residual components and performing statistical tests for stationarity and autocorrelation.

```python
def decompose_time_series(data, model='additive', figsize=(20, 8)):
    """
    Perform time series decomposition and visualize the components.
    
    Parameters:
    data (pandas.Series): Time series data to decompose.
    model (str): Decomposition model ('additive' or 'multiplicative') (default: 'additive').
    figsize (tuple): Figure size (default: (20, 8)).
    """
    try:
        rcParams['figure.figsize'] = figsize
        decomposition = sm.tsa.seasonal_decompose(data, model=model)
        fig = decomposition.plot()
        plt.show()
    except Exception as e:
        print(f"An error occurred during decomposition: {e}")
    
decompose_time_series(furniture_train)
```

The plot above clearly shows that the sales of furniture is unstable, along with its obvious seasonality.

# Time series forecasting with SARIMA (Seasonal Autoregressive Integrated Moving Average)

In the context of ARIMA (Autoregressive Integrated Moving Average) models, the parameters $p$, $d$, and $q$ represent the orders of autoregression, differencing, and moving average, respectively.

1. $p$ (Autoregressive Order): This parameter represents the number of lag observations included in the model. It indicates the number of previous time steps to consider for predicting the current value. A higher $p$ value means the model will consider more past observations.

2. $d$ (Differencing Order): This parameter represents the degree of differencing. It is the number of times the raw observations are differenced (i.e., subtracting an observation from the observation at the previous time step). Differencing is used to make the time series stationary, which is often necessary for modeling with ARIMA. A stationary time series has constant mean, variance, and autocorrelation over time.

3. $q$ (Moving Average Order): This parameter represents the size of the moving average window, i.e., the number of lag terms to include in the moving average part of the model. Each term represents the error of the model as a linear combination of error terms from previous observations. A higher $q$ value means the model will consider more past forecast errors.

**ARIMA (AutoRegressive Integrated Moving Average)**:
   - ARIMA models are a class of models that include autoregressive (AR), differencing (I), and moving average (MA) components.
   - The ARIMA model is specified by three main parameters: $p$, $d$, and $q$, which correspond to the autoregressive order, differencing order, and moving average order, respectively.
   - ARIMA models are suitable for non-seasonal time series data.

**SARIMA (Seasonal AutoRegressive Integrated Moving Average)**:
   - SARIMA models extend ARIMA models to account for seasonality in the time series data.
   - In addition to the parameters $p$, $d$, and $q$ for the non-seasonal components, SARIMA models include additional seasonal parameters: $P$, $D$, $Q$, and $s$, where $P$, $D$, and $Q$ are the seasonal autoregressive, differencing, and moving average orders, respectively, and $s$ is the length of the seasonal cycle (e.g., 12 for monthly data).
   - SARIMA models are suitable for time series data that exhibit seasonal patterns.

**SARIMAX (Seasonal AutoRegressive Integrated Moving Average with eXogenous variables)**:
   - SARIMAX models are an extension of SARIMA models that allow for the inclusion of exogenous variables, which are external factors that can influence the time series data.

Then, we perform a grid search over the parameter combinations generated earlier (pdq for non-seasonal parameters and seasonal_pdq for seasonal parameters) to fit SARIMA models to a the 4-year furniture sales time series data.

```python
import warnings
warnings.filterwarnings("ignore")

def grid_search_SARIMAX(data, p=range(0, 2), d=range(0, 2), q=range(0, 2), s=12):
    """
    Set up parameter combinations for a SARIMA model and perform a grid search over the parameter combinations to fit SARIMA models.
    """
    # Set up parameter combinations for a SARIMA model.
    pdq = list(itertools.product(p, d, q))
    seasonal_pdq = [(x[0], x[1], x[2], s) for x in pdq]
    
    # Obtain the name of the dataframe
    for name, obj in globals().items():
        if obj is data:
            dataframe_name = name
    
    # Store grid search value
    aic = {'param':[], 'param_seasonal':[], 'AIC':[]}
    
    # Perform grid search
    for param in pdq:
        for param_seasonal in seasonal_pdq:
            try:
                mod = sm.tsa.statespace.SARIMAX(data,
                                                order=param,
                                                seasonal_order=param_seasonal,
                                                enforce_stationarity=False, #mean, variance, and autocorrelation structure do not change over time
                                                enforce_invertibility=False) #past values of the series can be expressed as a linear combination of past error terms
                train_result = mod.fit(disp=False)
                aic['param'].append(param)
                aic['param_seasonal'].append(param_seasonal)
                aic['AIC'].append(train_result.aic) #Lower AIC values indicate a better trade-off between goodness of fit and model complexity
            except:
                continue
    
    idx = aic['AIC'].index(min(aic['AIC']))
    print('The SARIMAX{}x{} model fitted to the {} data has the lowest AIC value of {}'
          .format(aic['param'][idx], aic['param_seasonal'][idx], dataframe_name, aic['AIC'][idx]))
    
grid_search_SARIMAX(furniture_train)
```

# Fitting the SARIMAX model

```python
def fit_SARIMAX(data, pdq, seasonal_pdq):
    """
    Fit a SARIMAX model with predefined parameters on the training data and return the results object.
    """
    mod = sm.tsa.statespace.SARIMAX(data,
                                    order=pdq,
                                    seasonal_order=seasonal_pdq,
                                    enforce_stationarity=False,
                                    enforce_invertibility=False)
    results = mod.fit(disp=False)
    return results

pdq = (1, 1, 1)
seasonal_pdq = (0, 1, 1, 12)
results = fit_SARIMAX(furniture_train, pdq, seasonal_pdq)
print(results.summary().tables[1])
```

```python
def validate_prediction(test_data, results, data):
    """
    Make in-sample prediction on the test data and plot against the real observations.
    Also print out the Mean Squared Error and the Root Mean Squared Error.
    """
    pred = results.get_prediction(start=test_data.index[0], end=test_data.index[-1], dynamic=False)
    pred_ci = pred.conf_int()
    
    # plot the prediction against observation
    ax = data.plot(label='Observed')
    pred.predicted_mean.plot(ax=ax, label='Prediction', alpha=.7, figsize=(20, 8))
    ax.fill_between(pred_ci.index, pred_ci.iloc[:, 0], pred_ci.iloc[:, 1], color='k', alpha=.2)
    ax.set_xlabel('Date')
    ax.set_ylabel('Furniture Sales')
    plt.legend()
    plt.show()
    
    # Calculate the Mean Squared Error
    mse = ((pred.predicted_mean - test_data) ** 2).mean()
    sqrt_mse = np.sqrt(mse)
    print('The Mean Squared Error of our predictions is {}'.format(round(mse, 2)))
    print('The Root Mean Squared Error of our predictions is {}'.format(round(sqrt_mse, 2)))
    
    
validate_prediction(furniture_test, results, furniture)
```

The line plot is showing the observed values compared to the rolling forecast predictions. Overall, our predictions align with the true values very well, showing an upward trend starts from the beginning of the year and captured the seasonality toward the end of the year.

# Producing and visualizing forecasts

```python
def forecast(results, steps=100):
    """
    Produce out-of-sample forecasts and plot the forecasts against the original data.
    """
    pred_uc = results.get_forecast(steps=steps)
    pred_ci = pred_uc.conf_int()
    ax = furniture.plot(label='Observed', figsize=(20, 8))
    pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
    ax.fill_between(pred_ci.index,
                    pred_ci.iloc[:, 0],
                    pred_ci.iloc[:, 1], color='k', alpha=.25)
    ax.set_xlabel('Date')
    ax.set_ylabel('Furniture Sales')
    plt.legend()
    plt.show()

forecast(results)
```

The forecast commences from 2017 due to the fact that the model is fitted on training data that concludes at the end of 2016. Consequently, the 'results' of the model, encompass data only up to the end of 2016, initiating the forecasting process from 2017-01-01 onwards.

Our analysis effectively captures the seasonal patterns inherent in furniture sales data. As we extend our forecasting horizon, it's expected for our confidence in the predictions to diminish gradually. This is evidenced by the widening of the confidence intervals produced by our model, indicating increasing uncertainty as we project further into the future.

# Furniture vs. Office Supplies

The Office Supplies dataset is approximately 2 times larger than the Furniture dataset. 

To facilitate a comparison of sales between these two categories within the same time period, we will merge the two datasets into a single dataframe. Subsequently, we will plot the time series data for both categories on a single plot, enabling a visual comparison of their sales trends.

```python
def compare_sales(series_1, series_2):
    """
    merge sales data from two categories in grouped series form.
    """
    # Obtain the name of the dataframe
    for name, obj in globals().items():
        if obj is series_1:
            series_1_name = name
        if obj is series_2:
            series_2_name = name
    
    df1 = pd.DataFrame({'Order Date':series_1.index, 'Sales':series_1.values})
    df2 = pd.DataFrame({'Order Date': series_2.index, 'Sales': series_2.values})
    
    store = df1.merge(df2, how='inner', on='Order Date')
    store.rename(columns={'Sales_x': series_1_name+'_sales', 'Sales_y': series_2_name+'_sales'}, inplace=True)
    
    # Plot sales of two categories against each other
    label1 = series_1_name.capitalize().replace("_", " ")
    label2 = series_2_name.capitalize().replace("_", " ")
    plt.figure(figsize=(20, 8))
    plt.plot(store['Order Date'], store[series_1_name+'_sales'], 'b-', label = label1)
    plt.plot(store['Order Date'], store[series_2_name+'_sales'], 'r-', label = label2)
    plt.xlabel('Date'); plt.ylabel('Sales'); plt.title('Sales of {} and {}'.format(label1, label2))
    plt.legend();
    
    return store

store = compare_sales(furniture, office_supplies)
store.head()
```

We observe that sales of furniture and office supplies shared a similar seasonal pattern. Early of the year is the off season for both of the two categories. It seems summer time is quiet for office supplies too. In addition, average daily sales for furniture are higher than those of office supplies in most of the months. It is understandable, as the value of furniture should be much higher than those of office supplies. Occasionally, office supplies passed furniture on average daily sales.

# Time Series Modeling with Prophet

[Prophet](https://facebook.github.io/prophet/) is a procedure for forecasting time series data based on an additive model where non-linear trends are fit with yearly, weekly, and daily seasonality, plus holiday effects. It works best with time series that have strong seasonal effects and several seasons of historical data. Prophet is robust to missing data and shifts in the trend, and typically handles outliers well.

```python
from prophet import Prophet

def forecast_with_Prophet(data, periods, freq='MS'):
    """
    Use Prophet to make forecast on a time-series data.
    """
    for name, obj in globals().items():
        if obj is data:
            df_name = name.capitalize().replace("_", " ")
        
    # Rename columns to be accepted by Prophet. ds-(datestamp), y-(measurement we wish to forecast)
    df = pd.DataFrame({'ds':data.index, 'y':data.values})
    
    # Fit model to the data
    model = Prophet(interval_width=0.95)
    model.fit(df)
    
    # Make forecast
    forecast = model.make_future_dataframe(periods=periods, freq=freq)
    forecast = model.predict(forecast)
    
    # Plot the forecast
    plt.figure(figsize=(20, 8))
    model.plot(forecast, xlabel = 'Date', ylabel = 'Sales')
    plt.title(df_name+' Sales');
    
    return model, forecast
    
furniture_model, furniture_forecast = forecast_with_Prophet(furniture, 36)
office_supplies_model, office_supplies_forecast = forecast_with_Prophet(office_supplies, 36)
```
